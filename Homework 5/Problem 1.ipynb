## Part A


```python
import torch

# 1. Define the input data t_c (Celsius) as a tensor and t_u (unknown units) with noise
t_c = torch.arange(0.0, 100.0, 5.0) # Celsius from 0 to 95 in steps of 5
t_u = t_c * 10.0 + 9.0 + (torch.randn(len(t_c)) * 50.0) # Adding random noise

# 2. Define the non-linear model function
def model_non_linear(t_u, w2, w1, b):
    return w2 * t_u**2 + w1 * t_u + b

# 3. Define the mean squared error (MSE) loss function
def loss_fn(t_p, t_c):
    diff = (t_p - t_c)**2
    return diff.mean()

print("Input data t_c and t_u, non-linear model function, and MSE loss function defined.")
```

```python
import torch

# Initialize parameters w2, w1, and b
w2 = torch.tensor(1.0, requires_grad=True)
w1 = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)

# Define the training loop function
def training_loop(n_epochs, learning_rate, parameters, t_u, t_c, model_func, loss_func):
    for epoch in range(1, n_epochs + 1):
        # Calculate predicted values
        t_p = model_func(t_u, *parameters)

        # Calculate loss
        loss = loss_func(t_p, t_c)

        # Compute gradients
        loss.backward()

        # Update parameters
        with torch.no_grad():
            for param in parameters:
                param -= learning_rate * param.grad

        # Zero out gradients
        for param in parameters:
            param.grad.zero_()

        # Report loss every 500 epochs
        if epoch % 500 == 0 or epoch == 1:
            print(f"Epoch {epoch}, Loss {loss.item():.4f}")

    return parameters, loss.item()

print("Parameters initialized and generic training_loop function defined.")
```

## Part B


## Train with Learning Rate 0.1




```python
import torch

# Re-initialize parameters w2, w1, and b for training with LR 0.1
w2_0_1 = torch.tensor(1.0, requires_grad=True)
w1_0_1 = torch.tensor(1.0, requires_grad=True)
b_0_1 = torch.tensor(0.0, requires_grad=True)

# Call the training loop with n_epochs=5000, learning_rate=0.1
print("Starting training with learning rate 0.1...")
params_0_1, loss_0_1 = training_loop(
    n_epochs=5000,
    learning_rate=0.1,
    parameters=[w2_0_1, w1_0_1, b_0_1],
    t_u=t_u,
    t_c=t_c,
    model_func=model_non_linear,
    loss_func=loss_fn
)

print(f"Training with LR 0.1 finished. Final Loss: {loss_0_1:.4f}")
print(f"Final parameters for LR 0.1: w2={params_0_1[0].item():.4f}, w1={params_0_1[1].item():.4f}, b={params_0_1[2].item():.4f}")
```

## Identify Best Non-Linear Model and Train Linear Model


```python
import torch

# 1. Best non-linear model from previous training:
# The best non-linear model is the one trained with normalized t_u and LR 0.0001.
# Final Loss: 390.3954
# Final parameters: w2=14.1079, w1=19.0115, b=23.0304
# These values are stored in `params_normalized_0_0001` and `loss_normalized_0_0001`.

# 2. Define the linear model function
def model_linear(t_u, w, b):
    return w * t_u + b

# 3. Re-initialize parameters w and b for the linear model
w_linear = torch.tensor(1.0, requires_grad=True)
b_linear = torch.tensor(0.0, requires_grad=True)

# 4. Train the linear model using the training_loop function
print("\nStarting training for linear model with normalized t_u and learning rate 0.0001...")
params_linear, loss_linear = training_loop(
    n_epochs=5000,
    learning_rate=0.0001,
    parameters=[w_linear, b_linear],
    t_u=t_un, # Use normalized input data
    t_c=t_c,
    model_func=model_linear,
    loss_func=loss_fn
)

# 5. Print the final loss and parameters for the trained linear model
print(f"\nTraining for linear model finished. Final Loss: {loss_linear:.4f}")
print(f"Final parameters for linear model: w={params_linear[0].item():.4f}, b={params_linear[1].item():.4f}")
```
